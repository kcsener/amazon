```python
import pandas as pd
import numpy as np
import json
import requests
from bs4 import BeautifulSoup as BS
from selenium import webdriver
import time
import pyautogui
from selenium.webdriver.common.keys import Keys
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}
```

```python
url = 'https://www.amazon.com/s?k=nintendo+switch+console&ref=nb_sb_noss_1'
```

```python
response=requests.get(url, headers=headers)
```

```python
soup=BS(response.content, features="lxml")
```

```python
a = soup.select('a.a-link-normal.a-text-normal')
```

```python
l_pn = []
l_sb = []
l_p  = []
l_e  = []
l_c  = []
for i in a:
    url = 'https://www.amazon.com' + i['href']
    print(url)
    try:
        response=requests.get(url, headers=headers)
        soup=BS(response.content, features="lxml")
        try:
            product_name=soup.select("#productTitle")[0].get_text().strip()
            l_pn.append(product_name)
        except:
            l_pn.append(np.nan)
        try:
            sold_by=soup.select("#bylineInfo")[0].get_text().strip()
            l_sb.append(sold_by)
        except:
            l_sb.append(np.nan)
        try:
            price=soup.select("#priceblock_ourprice")[0].get_text().strip()
            l_p.append(price)
        except:
            l_p.append(np.nan)
        try:
            edition=soup.select(".selection")[1].get_text().strip()
            l_e.append(edition)
        except:
            l_e.append(np.nan)
        try:
            color=soup.select(".selection")[0].get_text().strip()
            l_c.append(color)
        except:
            l_c.append(np.nan)
    except:
        pass
```

```python
df = pd.DataFrame(list(zip(l_pn, l_sb, l_p, l_e, l_c)), 
               columns =['ProductName', 'SoldBy', 'Price', 'Edition', 'Color']) 
```

```python
df = df.drop_duplicates()
```
